# Final Visualization: Multivariate Time Series Anomaly Detection
This notebook visualizes the output CSV for all model abnormality scores and top contributing features.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# --- Exploratory Data Analysis (EDA) ---
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data_path = '../data/train.csv'
df = pd.read_csv(data_path, parse_dates=[0])
df = df.rename(columns={df.columns[0]: "timestamp"})
df = df.set_index("timestamp")

print("Data Info:")
df.info()
print("\nSummary Statistics:")
display(df.describe().T)
print("\nMissing Values per Column:")
print(df.isnull().sum())
print("\nCorrelation Matrix:")
display(df.corr())

# Histograms
for col in df.columns:
    plt.figure(figsize=(6, 4))
    df[col].hist(bins=30)
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()
    plt.close()
# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.tight_layout()
plt.show()
plt.close()
```

```python
# Load final output
df = pd.read_csv('../outputs/anomalies_output.csv')
df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')
df = df.set_index('timestamp')
df.head()
```

## Plot abnormality scores over time for each model

```python
models = ['abnormality_score_iforest', 'abnormality_score_pca', 'abnormality_score_lstm', 'abnormality_score_ensemble']
plt.figure(figsize=(15, 6))
for m in models:
    plt.plot(df.index, df[m], label=m)
plt.title('Abnormality Scores Over Time (All Models)')
plt.ylabel('Score (0-100)')
plt.legend()
plt.show()
```

## Overlay anomalies on a selected feature plot (Ensemble)

```python
feature = df.columns[0]  # Change to any feature you want to inspect
anomaly_mask = df['abnormality_score_ensemble'] > 80
fig, ax = plt.subplots(figsize=(15, 5))
ax.plot(df.index, df[feature], label=feature)
ax.scatter(df.index[anomaly_mask], df[feature][anomaly_mask], color='red', label='Anomaly', zorder=10)
ax.set_title(f'{feature} with Anomalies Overlay (Ensemble)')
ax.legend()
plt.show()
```

## Histogram of abnormality scores for each model

```python
plt.figure(figsize=(15, 6))
for m in models:
    plt.hist(df[m], bins=50, alpha=0.5, label=m)
plt.title('Histogram of Abnormality Scores (All Models)')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
